## Удалим незначащие строки

```python
# есть строки, где количество проданных билетов 0, возможно, это грузовые рейсы, их можно удалить
df = df[df.ticket_economy + df.ticket_comfort + df.ticket_business !=0]
```

```python
df.drop(['seat_comfort','ticket_comfort','sum_ticket_comfort'],axis=1,inplace = True)
```

## Найдем среднюю цену билета для каждого класса и потенциальную выручку

```python
# средняя цена билета эконом-класса
df['av_economy'] = df['sum_ticket_economy']/df['ticket_economy']
```

```python
#  потенциальная выручка для эконом-класса, если бы все места были заняты
# (средняя цена билета эконом-класса на этом рейсе, умноженная на общее количество мест эконом класса для этой модели самолета)
df['pot_sum_economy'] = df['av_economy'] * df['seat_economy']
```

```python
# средняя цена билета бизнес-класса
df['av_business'] = df['sum_ticket_business']/df['ticket_business']
```

```python
#  потенциальная выручка для бизнес-класса, если бы все места были заняты
# (средняя цена билета бизнес-класса на этом рейсе умноженная на общее количество мест бизнес-класса для этой модели самолета)
df['pot_sum_busines'] = df['av_business'] * df['seat_business']
```

```python
df.head()
```

##  Рассчитаем суммовые потери от незаполненности самолета

```python
# Это разница между потенциальной выручкой при полной занятости и фактической выручкой
df['lost_sum'] = df['pot_sum_economy'] - df['sum_ticket_economy'] + df['pot_sum_busines'] - df['sum_ticket_business']
```

```python
df
```

## Выделим рейсы с наибольшими суммовыми потерями от заполненности

```python
# топ 3 рейсов в Москву с наибольшими потерями
df[df.arrival_city == 'Moscow'].sort_values('lost_sum',ascending=False).head(3)
```

```python
# топ 3 рейсов в Белгород с наибольшими потерями
df[df.arrival_city == 'Belgorod'].sort_values('lost_sum',ascending=False).head(3)
```

## Рассчитаем процент заполненности самолета

```python
# рассчитаем процент заполненности самолета 
# как отношение количества занятых мест к максимальному количеству мест (количеству мест для данной модели самолета)
df['koeff_mest'] = (df['ticket_economy'] +df['ticket_business'])/(df['seat_economy'] + df['seat_business']) *100
```

## Рассчитаем стоимость топлива для оценки прибыльности исходя из расхода топлива

```python
# модели самолетов
df['model'].value_counts()
```

```python
# часовой расход топлива для Boeing 737-300 - 2600 кг/час
# часовой расход топлива для Sukhoi Superjet-100 - 1700 кг/час
# информация взята с сайта https://www.airlines-inform.ru/commercial-aircraft/boeing-737-300.html
def rashod(x):
    if x == 'Boeing 737-300': return 2600
    if x == 'Sukhoi Superjet-100': return 1700
```

```python
# применим функцию и заполним расход в датасете
df['rashod'] = df['model'].apply(rashod)
```

```python
# рассчитаем стоимость топлива, исходя из длительности
# стоимость топлива = длительность полета * часовой расход топлива * средняя цена за топливо
# за среднюю цену топлива по России возьмем 43 000 рублей за тонну, т.е. 43 руб за кг
# https://www.vedomosti.ru/business/articles/2017/11/28/743316-aviakerosin-podorozhal
df['sum_toplivo'] = df['actual_duration']/60 * df['rashod'] * 43
```

```python
 df
```

# Проведем оценку прибыльности исходя из расхода топлива

```python
# оценки прибыльности: стоимость билетов - стоимость топлива на рейс
df['prib'] = df['sum_ticket_economy'] + df['sum_ticket_business'] - df['sum_toplivo']
```

```python
# топ 3 неприбыльных рейса в Москву 
df[df.arrival_city == 'Moscow'].sort_values('prib',ascending=True).head(3)
```

```python
# топ 3 неприбыльных рейса в Белгород 
df[df.arrival_city == 'Belgorod'].sort_values('prib',ascending=True).head(3)
```
